*Questions 1-3*

Notes:
- 60mo = data from last 60 months
- pd.Series is meant to convert time series to a time series in Pandas format to allow us to complete a time series analysis.
-The resample function's syntax calls for a frequency. "BM corresponds to the last business day of each month (a combination of "B"=business day, "M"=month).

```python
data_AMZN=yf.download('AMZN',period="60mo")
NewTimeSeries_AMZN=data_AMZN['Adj Close']
NewSeries_AMZN=NewTimeSeries_AMZN.resample("BM").sum()
Pd_NewSeries_AMZN = pd.Series(NewSeries_AMZN)
Pd_NewSeries_AMZN
```

```python
data_IBM=yf.download('IBM',period="60mo")
NewTimeSeries_IBM=data_IBM['Adj Close']
NewSeries_IBM=NewTimeSeries_IBM.resample("BM").sum()
Pd_NewSeries_IBM = pd.Series(NewSeries_IBM)
Pd_NewSeries_IBM
```

```python
data_MMM=yf.download('MMM',period="60mo")
NewTimeSeries_MMM=data_MMM['Adj Close']
NewSeries_MMM=NewTimeSeries_MMM.resample("BM").sum()
Pd_NewSeries_MMM = pd.Series(NewSeries_MMM)
Pd_NewSeries_MMM
```

```python
data_META=yf.download('META',period="60mo")
NewTimeSeries_META=data_META['Adj Close']
NewSeries_META=NewTimeSeries_META.resample("BM").sum()
Pd_NewSeries_META = pd.Series(NewSeries_META)
Pd_NewSeries_META
```

*Question 4*

The autocorrelation plot shows that all four series are non-stationary. More importantly, however, is the fact that all series begin by having high strong autocorrelation within the first 10 months of analysis (as they surpassed the 99% CB, the dashed line), but as time approaches current month, all series become weaker in autocorrelation (shown as all four lines approaching 0). Beyond the initial 10 month period, the only two lags where there is strong autocorrelation present is around the 15th month, where META's correlation value reaches -0.5, and at around the 20th month, where AMZN's correlation value reaches around -0.375. In both cases, they surpassed the 99% CB.

```python
import matplotlib.pyplot as plt
aWNTimePlot = pd.plotting.autocorrelation_plot(Pd_NewSeries_AMZN, label='AMZN')
aWNTimePlot = pd.plotting.autocorrelation_plot(Pd_NewSeries_IBM, label='IBM')
aWNTimePlot = pd.plotting.autocorrelation_plot(Pd_NewSeries_MMM, label='MMM')
aWNTimePlot = pd.plotting.autocorrelation_plot(Pd_NewSeries_META, label='META')
aWNTimePlot.set_title("An Autocorrelation Graph of 4 Stocks")
aWNTimePlot.set_ylabel("Correlation")
aWNTimePlot.set_xlabel("Lag (End of Month)")
plt.show()
```

*Question 5*

```python
MonthlyReturns_AMZN=(Pd_NewSeries_AMZN / Pd_NewSeries_AMZN.shift(1) - 1).dropna()
MonthlyReturns_IBM=(Pd_NewSeries_IBM / Pd_NewSeries_IBM.shift(1) - 1).dropna()
MonthlyReturns_MMM=(Pd_NewSeries_MMM / Pd_NewSeries_MMM.shift(1) - 1).dropna()
MonthlyReturns_META=(Pd_NewSeries_META / Pd_NewSeries_META.shift(1) - 1).dropna()
MonthlyReturns_AMZN
```

```python
MonthlyReturns_META
```

Unlike with the previous autocorrelation plot, the monthly returns time series for each stocks do not surpass the 99% CB at any point. In other words, while there are certain lags where autocorrelation is stronger than in other lags (e.g., it is strongest at around the 17th month for AMZN, at around the 12th month for IBM and MMM, at around the 16th month for META), generally the autocrrelation is moderate-weak for most lags. Also a significant point of departure from the previous plot is the fact that autocorrelation does not approach 0 as time approaches current month. Instead, there appears to be a random fluctuation, indicating that the series are not completely random.

```python
fig, ax = plt.subplots()
aWNTimePlot_Returns = pd.plotting.autocorrelation_plot(MonthlyReturns_AMZN, label='AMZN')
aWNTimePlot_Returns = pd.plotting.autocorrelation_plot(MonthlyReturns_IBM, label='IBM')
aWNTimePlot_Returns = pd.plotting.autocorrelation_plot(MonthlyReturns_MMM, label='MMM')
aWNTimePlot_Returns = pd.plotting.autocorrelation_plot(MonthlyReturns_META, label='META')

aWNTimePlot_Returns.set_title("An Autocorrelation Graph of 4 Stocks-Returns")
aWNTimePlot_Returns.set_ylabel("Correlation")
aWNTimePlot_Returns.set_xlabel("Lag (End of Month)")
plt.show()
```
